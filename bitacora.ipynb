```python
import numpy as np
from decimal import Decimal as fp

#solo datos, cero calculos

balanza1 = [[1,1.0011,1.0011,1.0008],[2,2.0038,2.0038,2.0033],[5,5.0049,5.0048,5.0044],
            [10,10.0005,10.0001,9.9994],[20,20.0087,20.0081,20.0084],
            [50,49.9987,49.9984,49.9986],[100,100.0094,100.0099,100.0093]]

balanza4 = [[1,0.9834,0.9834,0.9834],[2,1.9684,1.9684,1.9684],[5,4.9168,4.9169,4.9169],
            [10,9.8244,9.8243,9.8242],[20,19.6568,19.6568,19.6569],
            [50,49.1206,49.1204,49.1204],[100,98.2542,98.2541,98.2540]]

balanzas_l5 = [[[2,2.0050],[5,5.0043],[10,9.9997],[20,20.0080],[50,49.9980],[100,100.0100]],#EObalanza1
              [[2,2.0035],[5,5.0051],[10,10.0000],[20,20.0090],[50,49.9980],[100,100.0100]],#EObalanza2
              [[2,2.0037],[5,5.0046],[10,9.9996],[20,20.0080],[50,49.9970],[100,100.0100]],#EObalanza3
              [[2,1.9702],[5,4.9171],[10,9.8247],[20,19.6580],[50,49.1220],[100,98.2570]],#EObalanza4
              [[2,2.0051],[5,5.0045],[10,9.9997],[20,20.0080],[50,49.9980],[100,100.0100]],#EObalanza5
              [[2,2.0039],[5,5.0046],[10,9.9998],[20,20.0080],[50,49.9970],[100,100.0100]],#EObalanza6
              [[2,2.0059],[5,5.0045],[10,9.9993],[20,20.0070],[50,49.9950],[100,100.0000]],#EObalanza7
              [[2,2.0036],[5,5.0047],[10,9.9995],[20,20.0080],[50,49.9980],[100,100.0100]]]#EOall 

monedas = [[3.8323,3.8322,3.8331],[3.8893,3.8404,3.8905],[3.8921,3.8935,3.8931],
           [3.9004,3.9013,3.9004],[3.9193,3.9307,3.9301]]

coin_diff = [19.4485,15.5152,11.7247,7.8313,3.9304]


```

```python
#calculos de las balanzas 1 y 4
promedios = [[],[]]
desviaciones = [[],[]]
errores_abs = [[],[]]
errores_rel = [[],[]]

promedios[0] = [np.average(balanza1[i][1:]) for i in range(len(balanza1))]
promedios[1] = [np.average(balanza4[i][1:]) for i in range(len(balanza4))]

desviaciones[0] = [np.std(balanza1[i][1:]) for i in range(len(balanza1))]
desviaciones[1] = [np.std(balanza4[i][1:]) for i in range(len(balanza4))]

errores_abs[0] = [abs(promedios[0][i] - balanza1[i][0]) for i in range(len(promedios[0]))]
errores_abs[1] = [abs(promedios[1][i] - balanza4[i][0]) for i in range(len(promedios[1]))]

for a in range(0,2):
    errores_rel[a] = [abs(errores_abs[a][i]/balanza1[i][0])*100 for i in range(len(errores_abs[a]))]

promedios = np.round(promedios, 4)
desviaciones = np.round(desviaciones, 4)
errores_abs = np.round(errores_abs, 4)
errores_rel = np.round(errores_rel, 4)
```

```python
#calculos de las balanzas del lab 5

errores_abs_2 = [[],[],[],[],[],[],[],[]] #one sublist per balance; inefficient, should be functionalized
errores_rel_2 = [[],[],[],[],[],[],[],[]]

for a in range(len(balanzas_l5)):
    errores_abs_2[a] = [abs(balanzas_l5[a][i][1] - balanzas_l5[a][i][0]) for i in range(len(balanzas_l5[a]))]
    
for a in range(len(errores_abs_2)):
    errores_rel_2[a] = [abs(errores_abs_2[a][i] / balanzas_l5[a][i][0])*100 for i in range(len(errores_abs_2[0]))]
    
errores_abs_2 = np.round(errores_abs_2, 4)
errores_rel_2 = np.round(errores_rel_2, 4)
```

```python
#calculos de monedas individuales

avg_coins = []
stdev_coins = []
err_coins = [[],[]] #1st list abs err, 2nd rel err

avg_coins = [(np.average(i[0:])) for i in monedas]
stdev_coins = [(np.std(i[0:])) for i in monedas]

for a in range(0,5):
    sum = [abs(i - avg_coins[a]) for i in monedas[a]]
    err_coins[0].append(np.sum(sum)/3)

err_coins[1] = [(err_coins[0][i]/avg_coins[i]*100) for i in range(len(err_coins[0]))]

avg_coins = np.round(avg_coins, 4)
stdev_coins = np.round(stdev_coins, 4)
err_coins = np.round(err_coins, 4)
```

```python
#calculos de monedas por diferencias

diff_coins = []
diff_coins_err = [[],[]] #1st list abs err, 2nd list rel err

diff_coins = [(coin_diff[i] - coin_diff[i+1]) for i in range(0,4)]
diff_coins.append(coin_diff[-1])
diff_coins_err[0] = abs(np.subtract(diff_coins, avg_coins))
diff_coins_err[1] = np.multiply(np.divide(diff_coins_err[0],avg_coins),100)

diff_coins = np.round(diff_coins, 4)
diff_coins_err = np.round(diff_coins_err, 4)
```
